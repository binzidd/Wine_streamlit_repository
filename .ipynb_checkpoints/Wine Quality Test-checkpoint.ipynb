```python
# Correlation matrix
def plotCorrelationMatrix(df, graphWidth):
    filename =df 
    df = df.dropna('columns') # drop columns with NaN
    df = df[[col for col in df if df[col].nunique() > 1]] # keep columns where there are more than 1 unique values
    if df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN or constant columns ({df.shape[1]}) is less than 2')
        return
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
 #   plt.title(f'Correlation Matrix for {filename}', fontsize=15)
    plt.show()
```

```python
plotCorrelationMatrix(wine,8)
```

### What do each of the measures mean?
   Input variables (based on physicochemical tests):<br>
   <font color='red'>associate each of these to something more relatable, e.g. 4= sweetness</font><br>
   1 - fixed acidity<br>
   2 - volatile acidity<br>
   3 - citric acid<br>
   4 - residual sugar <br>
   5 - chlorides  saltiness<br>
   6 - free sulfur dioxide  sulfur dioxide that is not bound to other chemicals in the solution and is free to react with other chemicals (see 7)<br>
   7 - total sulfur dioxide<br> common preservative to prevent microbial growth and oxidation / browning<br>
   8 - density<br>
   9 - pH - another acidity level measure <br>
   10 - sulphates<br> 
   11 - alcohol<br>
   Output variable (based on sensory data): <br><br>
   12 - quality (score between 0 and 10)




### Start with Decision Tree

Let audience pick up to <b> 5</b> features and run the model to determine MAE <br>
<b>Compete</b> for the lowest MAE!

```python
y=wine['quality']
features=['alcohol',
          'volatile acidity',
          'free sulfur dioxide',
          'residual sugar',
          'density',
          'pH'
         
    
         ]
x=wine[features]

train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 1234)

winemodel1 = DecisionTreeRegressor(random_state=1)
winemodel1.fit(train_x, train_y)

winemodel1.predict(val_x)
val_predictions1 = winemodel1.predict(val_x)
print('MAE=',mean_absolute_error(val_y, val_predictions1))
```

#### Visualise the decision tree

```python
## Redo model with lower max depth for visualisation
winemodel2 = DecisionTreeRegressor(random_state=2,max_depth=3)
winemodel2.fit(train_x, train_y)

winemodel2.predict(val_x)
val_predictions2 = winemodel2.predict(val_x)
print('MAE=',mean_absolute_error(val_y, val_predictions2))
```

```python
dot_data = tree.export_graphviz(winemodel2, out_file=None, 
                                feature_names=features,  
                                class_names=train_y,
                                filled=True)

# Draw graph
graph = graphviz.Source(dot_data, format="png") 
graph
```
